#### con if, else, elif , while  True , diccionario funciones

```python
# ==========================================
# PROYECTO UIII ACT 11
# IA AGENCIA DE MODELOS
# ==========================================

print("======================================")
print("👠 IA DE AGENCIA DE MODELOS 👠")
print("======================================")

# Lista de diccionarios vacía
modelos = []

# FUNCIÓN AGREGAR MODELO
def agregar_modelo():

    nombre = input("Nombre del modelo: ")
    estatura = input("Estatura: ")
    tipo = input("Tipo de modelaje (pasarela/editorial): ")
    marcas = input("Marcas representadas: ")
    portafolio = input("Portafolio: ")
    viaje = input("Disponibilidad de viaje: ")

    modelo = {
        "Nombre": nombre,
        "Estatura": estatura,
        "Tipo": tipo,
        "Marcas": marcas,
        "Portafolio": portafolio,
        "Viaje": viaje
    }

    modelos.append(modelo)

    print("✅ Modelo agregado correctamente")


# FUNCIÓN MOSTRAR MODELOS
def mostrar_modelos():

    if modelos == []:

        print("❌ No hay modelos registrados")

    else:

        print("\n================ TABLA DE MODELOS ================")

        print("Nombre\tEstatura\tTipo\t\tMarcas\t\tViaje")

        print("--------------------------------------------------")

        for m in modelos:

            print(m["Nombre"], "\t",
                  m["Estatura"], "\t\t",
                  m["Tipo"], "\t",
                  m["Marcas"], "\t",
                  m["Viaje"])


# FUNCIÓN IA
def recomendacion_ia():

    print("\n🤖 RECOMENDACIÓN IA")

    estilo = input("¿Qué tipo de modelaje buscas?: ")

    if estilo == "pasarela":

        print("✅ IA recomienda modelos altos y elegantes")

    elif estilo == "editorial":

        print("✅ IA recomienda modelos creativos y fotogénicos")

    else:

        print("❌ Tipo de modelaje no válido")


# FUNCIÓN ELIMINAR
def eliminar_modelo():

    nombre = input("Nombre del modelo a eliminar: ")

    encontrado = False

    for m in modelos:

        if m["Nombre"] == nombre:

            modelos.remove(m)

            encontrado = True

            print("✅ Modelo eliminado")

            break

    if encontrado == False:

        print("❌ Modelo no encontrado")


# MENÚ PRINCIPAL
while True:

    print("\n=========== MENÚ ===========")
    print("1. Agregar modelo")
    print("2. Ver modelos")
    print("3. Recomendación IA")
    print("4. Eliminar modelo")
    print("5. Salir")

    opcion = input("Selecciona una opción: ")

    # OPCIONES
    if opcion == "1":

        agregar_modelo()

    elif opcion == "2":

        mostrar_modelos()

    elif opcion == "3":

        recomendacion_ia()

    elif opcion == "4":

        eliminar_modelo()

    elif opcion == "5":

        print("\n👋 Gracias por usar la IA de Agencia")
        break

    else:

        print("❌ Opción inválida")
        print("Programa desarrollado por Paola Silveyra NL = 55")
```
